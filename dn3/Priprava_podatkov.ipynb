Ukvarjali se bomo z učenjem iz kompleksnih podatkovnih tipov, natančneje iz devetih tabel, ki opisujejo ligo NBA.

Ta Jupyter notebook poskrbi za pretvorbo podatkov iz tabel v primerno obliko, ki jo lahko sprejme paket re3py. Notebook prebere datoteke v mapi *dn3* (torej je pot oblike *dn3/datoteka.csv*). Nalogo oddajam brez te datoteke, zato jo je potrebno pred poganjanjem dodati v isto mapo kjer se nahaja *Priprava_podatkov.ipynb*.

```python
# imports
import pandas as pd
import math
```

```python
# preberemo tabele iz .csv datotek
teams = pd.read_csv("dn3/teams.csv", sep="\t") 

allstar = pd.read_csv("dn3/allstar.csv", sep="\t") 
coach_season = pd.read_csv("dn3/coach_season.csv", sep="\t") 
# coaches = pd.read_csv("dn3/coaches.csv", sep="\t") # nepotrebna tabela
draft = pd.read_csv("dn3/draft.csv", sep="\t")
players = pd.read_csv("dn3/players.csv", sep="\t") 
playoffs = pd.read_csv("dn3/playoffs.csv", sep="\t") 
regular_season = pd.read_csv("dn3/regular_season.csv", sep="\t") 
team_season = pd.read_csv("dn3/team_season.csv", sep="\t") 

# imena tabel povežemo z spremenljivkami kjer so shranjene kot pandas dataframe
tabele = [(allstar, 'allstar'), (coach_season, 'coach_season'), (draft, 'draft'), 
          (players, 'players'), (playoffs, 'playoffs'), (regular_season, 'regular_season'), 
          (team_season, 'team_season'), (teams, 'teams')]

tabele_player = [(allstar, 'allstar'), (draft, 'draft'), (players, 'players'), 
                 (playoffs, 'playoffs'), (regular_season, 'regular_season')]

tabele_team = [(draft, 'draft'), (coach_season, 'ciach_Season'), (playoffs, 'playoffs'), 
               (regular_season, 'regular_season'), (team_season, 'team_season'), (teams, 'teams')]

tabele_coach = [(coach_season, 'coach_season')]

# tipi atributov
povezovalni_atributi = ['playerid', 'coachid', 'team']
ciljni_atrbut = ['league']
nominalni_atributi = ['firstname', 'lastname', 'location', 'name', 'conf', 'position']
```

**(1)** Najprej konstruiramo datoteko basket_opis.txt. Ta datoteka vsebuje vse možne pare odebeljenih ter neodebeljenih argumentov. 

```python
IME1 = 'basket_opis.txt'

# konstruiramo podatke v pravi obliki
with open(IME1, 'w') as f:
    # Allstar
    for k in allstar.keys():
        if k != 'playerid':
            for i in range(0, len(allstar['playerid'])):
                if allstar[k][i] == allstar[k][i]:
                    vrstica = "allstar_"+str(k)+"("+str(allstar['playerid'][i])+", "+str(allstar[k][i])+")"
                    f.write(vrstica)
                    f.write("\n")
                    
    # Coach_season
    for k in coach_season.keys():
        if k != 'coachid':
            for i in range(0, len(coach_season['coachid'])):
                vrstica = "coach_season_"+str(k)+"("+str(coach_season['coachid'][i])+", "+str(coach_season[k][i])+")"
                f.write(vrstica)
                f.write("\n")
                
    # Draft
    for k in draft.keys():
        if k != 'team':
            for i in range(0, len(draft['team'])):
                besedilo1 = draft[k][i]
                besedilo = "".join(x for x in str(besedilo1) if x not in "û–:,-_ ';&?!=+*$()/")
                vrstica = "draft_"+str(k)+"("+str(draft['team'][i])+", "+str(besedilo)+")"
                f.write(vrstica)
                f.write("\n")
                             
    # Players
    for k in players.keys():
        if k != 'playerid':
            for i in range(0, len(players['playerid'])):
                if players[k][i] == players[k][i]:
                    if players[k][i] != '?':
                        besedilo1 = players[k][i]
                        besedilo = "".join(x for x in str(besedilo1) if x not in "û–:,-_ ';&?!=+*$()/")
                        vrstica = "players_"+str(k)+"("+str(players['playerid'][i])+", "+str(besedilo)+")"
                        f.write(vrstica)
                        f.write("\n")
            
    # Playoffs
    for k in playoffs.keys():
        if k != 'playerid':
            for i in range(0, len(playoffs['playerid'])):
                vrstica = "playoffs_"+str(k)+"("+str(playoffs['playerid'][i])+", "+str(playoffs[k][i])+")"
                f.write(vrstica)
                f.write("\n")
            
    # Regular_season
    for k in regular_season.keys():
        if k != 'playerid':
            for i in range(0, len(regular_season['playerid'])):
                if regular_season[k][i] == regular_season[k][i]:
                    vrstica = "regular_season_"+str(k)+"("+str(regular_season['playerid'][i])+", "+str(regular_season[k][i])+")"
                    f.write(vrstica)
                    f.write("\n")
            
    # Team_season
    for k in team_season.keys():
        if k != 'team':
            for i in range(0, len(team_season['team'])):
                vrstica = "team_season_"+str(k)+"("+str(team_season['team'][i])+", "+str(team_season[k][i])+")"
                f.write(vrstica)
                f.write("\n")
                    
    # Teams
    for k in teams.keys():
        if k != 'team' and k != 'league':
            for i in range(0, len(teams['team'])):
                if teams[k][i] == teams[k][i]:
                    vrstica = "teams_"+str(k)+"("+str(teams['team'][i])+", "+str(teams[k][i])+")"
                    f.write(vrstica)
                    f.write("\n")
            
```

**(2)** Konstruiramo datoteko basket_ciljna.txt. Ta datotek vsebuje vse pare začetne in ciljne spremenljivke. Ta datetoka nam torej pove, kaj moramo napovedovati. 

```python
IME2 = 'basket_ciljna.txt'

# konstruiramo imena ekip in njihove ciljne spremenljivke
with open(IME2, 'w') as g:
    for i in range(0, len(teams['league'])):
        ekipa = teams['team'][i]
        liga = teams['league'][i]
        vrstica = "teams_league("+str(ekipa)+", "+str(liga)+")"
        g.write(vrstica)
        g.write("\n")
```

**(3)** Nazadnje konstruiramo še datoteko basket.s. Ta datoteka pa vsebuje vse možne
- relacije ([Relations]),
- agregate ([Aggregates]) in 
- teste ([AtomTests]).

V našem primeru bomo našteli vse možne teste in nato ročno zakomentirali tiste teste, za katere vemo da so škodljivi za model (player_firstname, player_lastname, ...) in tiste teste, ki jih želimo preveriti, če pomagajo modelu ali ne.

```python
IME3 = 'basket.s'
                     
#konstruitamo še datoteko basket.s
with open(IME3, 'w') as h:
    # relacije
    h.write("[Relations]\n")
    
    # postavimo ciljnega
    h.write("teams_league(TeamId, nominal00)\n")
    
    stevec = 1
    
    # Allstar
    for k in allstar.keys():
        if k != 'playerid':
            if k in nominalni_atributi:
                vrstica = "allstar_"+str(k)+"(PlayerId, nominal"+str(stevec)+")"
            else:
                vrstica = "allstar_"+str(k)+"(PlayerId, numeric"+str(stevec)+")"

            h.write(vrstica)
            h.write("\n")
            stevec += 1
            
    # Coach_season
    for k in coach_season.keys():
        if k != 'coachid':
            if k == 'team':
                vrstica = "coach_season_"+str(k)+"(CoachId, TeamId)"
                stevec -= 1
            elif k in nominalni_atributi:
                vrstica = "coach_season_"+str(k)+"(CoachId, nominal"+str(stevec)+")"
            else:
                vrstica = "coach_season_"+str(k)+"(CoachId, numeric"+str(stevec)+")"
                
            h.write(vrstica)
            h.write("\n")
            stevec += 1
            
    # Draft
    for k in draft.keys():
        if k != 'team':
            if k == 'playerid':
                vrstica = "draft_playerid(TeamId, nominal"+str(stevec)+")"
            elif k == 'draft_from':
                vrstica = "draft_"+str(k)+"(TeamId, SchoolTeam)"
                stevec -= 1
            elif k in nominalni_atributi:
                vrstica = "draft_"+str(k)+"(TeamId, nominal"+str(stevec)+")"
            else:
                vrstica = "draft_"+str(k)+"(TeamId, numeric"+str(stevec)+")"
                
            h.write(vrstica)
            h.write("\n")
            stevec += 1
    
    # Players
    for k in players.keys():
        if k != 'playerid':
            if k == 'college':
                vrstica = "players_"+str(k)+"(PlayerId, SchoolTeam)"
                stevec -= 1
            elif k in nominalni_atributi:
                vrstica = "players_"+str(k)+"(PlayerId, nominal"+str(stevec)+")"
            else:
                vrstica = "players_"+str(k)+"(PlayerId, numeric"+str(stevec)+")"

            h.write(vrstica)
            h.write("\n")
            stevec += 1
            
    # Playoffs
    for k in playoffs.keys():
        if k != 'playerid':
            if k == 'team':
                vrstica = "playoffs_"+str(k)+"(PlayerId, TeamId)"
                stevec -= 1
            elif k in nominalni_atributi:
                vrstica = "playoffs_"+str(k)+"(PlayerId, nominal"+str(stevec)+")"
            else:
                vrstica = "playoffs_"+str(k)+"(PlayerId, numeric"+str(stevec)+")"

            h.write(vrstica)
            h.write("\n")
            stevec += 1
            
    # Regular_season
    for k in regular_season.keys():
        if k != 'playerid':
            if k == 'team':
                vrstica = "regular_season_"+str(k)+"(PlayerId, TeamId)"
                stevec -= 1
            elif k in nominalni_atributi:
                vrstica = "regular_season_"+str(k)+"(PlayerId, nominal"+str(stevec)+")"
            else:
                vrstica = "regular_season_"+str(k)+"(PlayerId, numeric"+str(stevec)+")"

            h.write(vrstica)
            h.write("\n")
            stevec += 1
            
    # Team_season
    for k in team_season.keys():
        if k != 'team':
            if k in nominalni_atributi:
                vrstica = "team_season_"+str(k)+"(TeamId, nominal"+str(stevec)+")"
            else:
                vrstica = "team_season_"+str(k)+"(TeamId, numeric"+str(stevec)+")"

            h.write(vrstica)
            h.write("\n")
            stevec += 1
    
    # Teams
    for k in teams.keys():
        if k != 'team' and k != 'league':
            vrstica = "teams_"+str(k)+"(TeamId, nominal"+str(stevec)+")"

            h.write(vrstica)
            h.write("\n")
            stevec += 1
    
            
#################################################################################################  
#################################################################################################      
#################################################################################################      
    # dovoljeni agregati
    agregati = ['count', 'sum']
    h.write("[Aggregates]\n")
    for a in range(0, len(agregati)):
        h.write(agregati[a])
        h.write("\n")
    
#################################################################################################  
#################################################################################################      
################################################################################################# 
    # dovoljeni testi
    # katere teste bomo zakomentirali "//test(old, new)" se potem odločimo ročno!
    h.write("[AtomTests]\n")
        # Allstar
    for k in allstar.keys():
        if k != 'playerid':
            vrstica = "allstar_"+str(k)+"(old, new)"

            h.write(vrstica)
            h.write("\n")
            
    # Coach_season
    for k in coach_season.keys():
        if k != 'coachid':
            vrstica = "coach_season_"+str(k)+"(old, new)"
                
            h.write(vrstica)
            h.write("\n")
            
    # Draft
    for k in draft.keys():
        if k != 'team':
            vrstica = "draft_"+str(k)+"(old, new)"
                
            h.write(vrstica)
            h.write("\n")
    
    # Players
    for k in players.keys():
        if k != 'playerid':
            vrstica = "players_"+str(k)+"(old, new)"

            h.write(vrstica)
            h.write("\n")
            
    # Playoffs
    for k in playoffs.keys():
        if k != 'playerid':
            vrstica = "playoffs_"+str(k)+"(old, new)"

            h.write(vrstica)
            h.write("\n")
            
    # Regular_season
    for k in regular_season.keys():
        if k != 'playerid':
            vrstica = "regular_season_"+str(k)+"(old, new)"

            h.write(vrstica)
            h.write("\n")
            
    # Team_season
    for k in team_season.keys():
        if k != 'team':
            vrstica = "team_season_"+str(k)+"(old, new)"

            h.write(vrstica)
            h.write("\n")
    
    # dodamo še lokacijo
    h.write("teams_location(old, new)")
    h.write("\n")
       
        
# Na tem mestu ročno zakomentiramo neprimerne teste.
```

Ko so datoteke skonstruirane lahko ročno komentiramo teste, ki jih ne želimo. Ročno lahko dodajamo tudi nove teste. Npr. test oblike datoteka_atribut(new, old).
